```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float, exposure

image_full = img_as_float(data.camera())
image_low_contrast = exposure.rescale_intensity(image_full, out_range=(0.1,0.4))
image_stretched = exposure.rescale_intensity(image_low_contrast, in_range='image', out_range=(0,1))
image_equalized = exposure.equalize_hist(image_low_contrast)
fig, axes = plt.subplots(1,3, figsize=(15,5))
axes[0].imshow(image_low_contrast, cmap='gray'); axes[0].axis('off')
axes[1].imshow(image_stretched, cmap='gray'); axes[1].axis('off')
axes[2].imshow(image_equalized, cmap='gray'); axes[2].axis('off')
plt.show()
```
