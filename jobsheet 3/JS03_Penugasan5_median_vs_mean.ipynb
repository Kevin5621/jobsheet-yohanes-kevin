```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float, filters, util, morphology
from scipy import ndimage as ndi

gray_uint8 = data.camera()
gray_img = img_as_float(gray_uint8)
gray_noisy = util.random_noise(gray_img, mode='s&p', amount=0.05)
kernel_mean = np.ones((5,5))/25
gray_mean = ndi.convolve(gray_noisy, kernel_mean, mode='reflect')
gray_gauss = filters.gaussian(gray_noisy, sigma=1.5)
disk3 = morphology.disk(3)
gray_median = filters.median(gray_noisy, disk3)
fig, axes = plt.subplots(1,4, figsize=(20,5))
axes[0].imshow(gray_img, cmap='gray'); axes[0].axis('off')
axes[1].imshow(gray_noisy, cmap='gray'); axes[1].axis('off')
axes[2].imshow(gray_mean, cmap='gray'); axes[2].axis('off')
axes[3].imshow(gray_median, cmap='gray'); axes[3].axis('off')
plt.show()
```
