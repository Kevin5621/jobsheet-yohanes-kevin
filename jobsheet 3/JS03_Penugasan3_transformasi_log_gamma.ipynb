```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float, color, exposure

camera_uint8 = data.camera()
a_stronaut_uint8 = data.astronaut()
astro_gray = color.rgb2gray(a_stronaut_uint8)
camera = img_as_float(camera_uint8)
astro = astro_gray
log_camera = np.log1p(camera)/np.log1p(1.0)
fig, axes = plt.subplots(1,2, figsize=(10,4))
axes[0].imshow(camera, cmap='gray'); axes[0].axis('off')
axes[1].imshow(log_camera, cmap='gray'); axes[1].axis('off')
plt.show()
```
