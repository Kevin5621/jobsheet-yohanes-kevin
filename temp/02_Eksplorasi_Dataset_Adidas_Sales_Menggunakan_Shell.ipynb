### 5. Tampilkan Top 10 Sales Terbesar

**Hinst:**
1. Gunakan perintah dengan pola: **tail** .. | **sort** .. | **head**
2. Untuk mengexclude header, pada bagian tail gunakan opsi **-n +2**
3. Gunakan field **Total Sales** yang berada pada kolom ke 10
4. Untuk mengurutkan kolom ke 10 desc, gunakan tipe kolom **numerik**, sebagaimana dijelaskan pada **slide 32**

```python
# Tampilkan Top 10 Sales Terbesar
# tail -n +2: exclude header
# sort -t$'\t' -k10 -nr: sort by column 10 (Total Sales) in descending numeric order
# head -10: display top 10 records
! tail -n +2 data/adidas-sales-dataset.tsv | sort -t$'\t' -k10 -nr 2>/dev/null | head -10
```

Tampilkan hanya field City, Product, dan Total Sales

**Hinst:**
1. Gunakan kolom ke 6, 7, dan 10
2. Tambahkan perintah awk pada kode sebelumnya
3. Penggunaan awk dijelaskan pada **slide 34**

```python
# Tampilkan hanya field City, Product, dan Total Sales
# Kolom: 6 (City), 7 (Product), 10 (Total Sales)
! tail -n +2 data/adidas-sales-dataset.tsv | sort -t$'\t' -k10 -nr 2>/dev/null | head -10 | awk -F'\t' '{print $6, $7, $10}'
```

Tampilkan hanya field City, Product, dan Total Sales.

Ubah delimiter menjadi pipe "|"

**Hinst:**
1. Gunakan opsi **OFS** pada awk, sebagaimana dijelaskan pada **slide 25**


```python
# Tampilkan dengan pipe "|" delimiter
# Gunakan OFS (Output Field Separator) pada awk
! tail -n +2 data/adidas-sales-dataset.tsv | sort -t$'\t' -k10 -nr 2>/dev/null | head -10 | awk -F'\t' 'BEGIN {OFS="|"} {print $6, $7, $10}'
```

### 6. Tampilkan 10 produk dengan operating profit terkecil

Tampilkan fied **Product, Units Sold, Total Sales, dan Operating Profit**

Hinst:
1. Gunakan dan modifikasi kode sebelumnya
2. Output kolom ke 7, 9, 10,  dan 11

```python
# Tampilkan 10 produk dengan operating profit terkecil
# Kolom: 7 (Product), 9 (Units Sold), 10 (Total Sales), 11 (Operating Profit)
# Sort by column 11 (Operating Profit) in ascending numeric order
! tail -n +2 data/adidas-sales-dataset.tsv | sort -t$'\t' -k11 -n 2>/dev/null | head -10 | awk -F'\t' 'BEGIN {OFS="|"} {print $7, $9, $10, $11}'
```

### 7. Ratailer mana saja yang menjual produk Adidas?

**Hinst:**
1. Gunakan perintah: **awk** .. | **tail** .. | **sort** | **uniq**
2. Retailer berada pada kolom ke 1
3. Gunakan tail untuk mengexclude header

```python
# Retailer mana saja yang menjual produk Adidas?
# Kolom 1 berisi Retailer, gunakan awk | tail | sort | uniq
! awk -F'\t' '{print $1}' data/adidas-sales-dataset.tsv | tail -n +2 | sort | uniq
```

### 8. Produk apa saja yang dijual oleh Adidas?

Hinst:
1. Produk berada pada kolom ke 7

```python
# Produk apa saja yang dijual oleh Adidas?
# Kolom 7 berisi Product
! awk -F'\t' '{print $7}' data/adidas-sales-dataset.tsv | tail -n +2 | sort | uniq
```
